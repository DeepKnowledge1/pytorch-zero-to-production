<div style="background: rgba(96, 165, 250, 0.1); border-left: 5px solid #60a5fa; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #60a5fa; font-size: 24px; margin-top: 0;">1Ô∏è‚É£ NumPy ‚Üí PyTorch Tensor</h3>
  <p style="color: #4405f1ff; font-size: 18px;">Basic conversion (shares memory)</p>
</div>


```python
import numpy as np
import torch

np_array = np.array([1, 2, 3])
tensor_shared = torch.from_numpy(np_array)

tensor_shared
```

```python
np_array[0]=100

tensor_shared
```

<div style="background: rgba(239, 68, 68, 0.1); border-left: 5px solid #ef4444; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <p style="color: #b91c1c; font-size: 18px; margin: 0;">
    ‚ö†Ô∏è Shares memory ‚Äî changes affect both. NumPy array must be CPU & contiguous.
  </p>
</div>


<div style="background: rgba(96, 165, 250, 0.1); border-left: 5px solid #60a5fa; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #60a5fa; font-size: 22px;">Force copy (no shared memory)</h3>
</div>


```python
tensor_copy = torch.tensor(np_array)
tensor_copy
```

```python
np_array[0]=20

tensor_copy
```

<div style="background: rgba(96, 165, 250, 0.1); border-left: 5px solid #60a5fa; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #60a5fa; font-size: 22px;">Specify dtype</h3>
</div>


```python
torch.from_numpy(np_array).float()
torch.from_numpy(np_array).long()
torch.tensor(np_array, dtype=torch.float32)
```

<div style="background: rgba(96, 165, 250, 0.1); border-left: 5px solid #60a5fa; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #60a5fa; font-size: 24px;">2Ô∏è‚É£ PyTorch Tensor ‚Üí NumPy</h3>
</div>


```python
np_basic = tensor_shared.numpy()
np_basic
```

<div style="background: rgba(239, 68, 68, 0.1); border-left: 5px solid #ef4444; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <p style="color: #b91c1c; font-size: 18px;">
    Requirements: tensor must be on CPU and must not require gradients.
  </p>
</div>


```python
# Safe conversions
tensor_shared.detach().numpy()
tensor_shared.detach().cpu().numpy()
```

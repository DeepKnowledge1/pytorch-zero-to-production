<div style="background: rgba(59, 130, 246, 0.1); border-left: 5px solid #3b82f6; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #3b82f6; font-size: 24px;">1Ô∏è‚É£ Tensor Creation</h3>
</div>

```python
import torch

torch.tensor([1, 2, 3])      # from Python list
torch.zeros((2, 3))          # all zeros
torch.ones((3, 3))           # all ones
torch.empty((2, 2))          # uninitialized
torch.randn((3, 3))          # random normal
torch.arange(0, 10, 2)       # range
torch.linspace(0, 1, 5)      # evenly spaced values
```

<div style="background: rgba(59, 130, 246, 0.1); border-left: 5px solid #3b82f6; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #3b82f6; font-size: 24px;">2Ô∏è‚É£ Shape & Reshaping</h3>
</div>

```python
x = torch.randn(2, 3, 4)
x.shape                       # get shape
x.view(2, -1)                  # reshape
x.reshape(2, 12)               # another reshape
x.transpose(0, 1)              # swap dims
x.permute(2, 0, 1)             # reorder dims
x.squeeze()                     # remove dims of 1
x.unsqueeze(0)                  # add dim


# shape : 1,5,3 >> 5,3
```

<div style="background: rgba(59, 130, 246, 0.1); border-left: 5px solid #3b82f6; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #3b82f6; font-size: 24px;">3Ô∏è‚É£ Indexing & Slicing</h3>
</div>

```python
x = torch.arange(10)
x[0]                            # single element
x[2:6]                          # slicing
x[x > 5]                        # conditional
x[[1, 3, 5]]                    # fancy indexing
```

<div style="background: rgba(59, 130, 246, 0.1); border-left: 5px solid #3b82f6; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #3b82f6; font-size: 24px;">4Ô∏è‚É£ Math Operations</h3>
</div>

```python
a = torch.randn(3, 3)
b = torch.randn(3, 3)

a + b
a - b
a * b                           # element-wise
a / b
torch.matmul(a, b)              # matrix multiplication
a.sum()
a.mean()
a.max()
torch.sqrt(a)
torch.exp(a)
```

<div style="background: rgba(234, 179, 8, 0.1); border-left: 5px solid #eab308; padding: 15px; border-radius: 0 10px 10px 0; margin-bottom: 20px;">
  <h3 style="color: #eab308; font-size: 24px;">5Ô∏è‚É£ Device Management</h3>
</div>

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
x = torch.randn(3, 3).to(device)
x.device
x.cpu()                         # move to CPU
x.cuda() if torch.cuda.is_available() else x   # move to GPU
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**3 + 2*x
y.backward()
x.grad
```
